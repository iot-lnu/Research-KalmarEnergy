```python
from tabulate import tabulate

# description = modified_final_df.describe(include='all')
columns_of_interest = ['Power_Consumption', 'One_Day_Power_NaN', 'Dew Point Temperature_P39', 'Relative Humidity_P6', 'Wind Speed_P4', 'Price']
description = modified_final_df[columns_of_interest].describe()
print(tabulate(description, headers='keys', tablefmt='psql'))
```

```python
# Find the record with the minimum Price value (-60.04 euro/MWh is on 2023-07-16 at 12:00:00)
min_price_record = modified_final_df[modified_final_df['Price'] == modified_final_df['Price'].min()]

# Display the record(s) with the minimum price value
print(min_price_record)
```

```python
# Find the record with the maximum Price value (799.97 euro/MWh is on 2022-08-30 at 17:00:00))
max_price_record = modified_final_df[modified_final_df['Price'] == modified_final_df['Price'].max()]

# Display the record(s) with the maximum price value
print(max_price_record)
```

```python
# Count the number of records with a negative price
negative_price_count = (modified_final_df['Price'] < 0).sum()

print(f'Number of records with a negative price: {negative_price_count}')
```

```python
# Remove records with a negative price.
modified_final_df = modified_final_df[modified_final_df['Price'] >= 0]

# Save the cleaned dataset to a new CSV file, if needed.
modified_final_df.to_csv('data/modified_final_df_negativePricesRemoved.csv', index=False)

print("Records with negative prices have been removed.")
```

# 3 Sigma rule for removing records with very high (not logical) price values. NOT SURE IF I SHOULD APPLY IT OR SHOULD I APPLY IT ONLY ON RESIDENTIAL RECORDS?

```python
 #Calculate the mean and standard deviation of the prices.
mean_price = modified_final_df['Price'].mean()
std_price = modified_final_df['Price'].std()

# Define the upper threshold using the 3-sigma rule.
upper_limit = mean_price + (3 * std_price)

# Filter out the records where the price is greater than the upper limit.
modified_final_df = modified_final_df[modified_final_df['Price'] <= upper_limit]

# Save the filtered dataset to a new CSV file, if needed.
modified_final_df.to_csv('data/modified_final_df_negativePricesRemovedAnd3SigmaApplied.csv', index=False)

print(f"Records with prices higher than 3 standard deviations from the mean have been removed.")
```
