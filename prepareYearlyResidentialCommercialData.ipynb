```python
import pandas as pd

# Load the data
filename = 'data/modified_final_df.csv'
df = pd.read_csv(filename, sep=',', index_col=0)

df.shape[0]

# Split the data into commercial and residential dataframes
residential_df = df[df['ISPRIVATEPERSON'] == 'Ja']
commercial_df = df[df['ISPRIVATEPERSON'] == 'Nej']

# Ensure 'DateTime' is a datetime column
commercial_df['DateTime'] = pd.to_datetime(commercial_df['DateTime'])
residential_df['DateTime'] = pd.to_datetime(residential_df['DateTime'])

# Get the unique years from the 'DateTime' column
years = commercial_df['DateTime'].dt.year.unique()

# Loop through each year and create separate DataFrames and CSVs
for year in years:
    # Filter by year for commercial
    commercial_year_df = commercial_df[commercial_df['DateTime'].dt.year == year]
    commercial_csv_filename = f'data/yearlyData/commercial_df_{year}.csv'
    commercial_year_df.to_csv(commercial_csv_filename, sep=',')
    
    # Filter by year for residential
    residential_year_df = residential_df[residential_df['DateTime'].dt.year == year]
    residential_csv_filename = f'data/yearlyData/residential_df_{year}.csv'
    residential_year_df.to_csv(residential_csv_filename, sep=',')
```

# Always 2 steps before modeling using timeseries 1) Convert 'DateTime' to datetime if it's not already 2) Set 'DateTime' as index

```python
# Convert 'DateTime' to datetime if it's not already
df['DateTime'] = pd.to_datetime(df['DateTime'])
```
