```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data
df = pd.read_csv('data/yearlyData/commercial_df_2020.csv')

# Ensure that DateTime is a datetime type
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Set DateTime as the index
df.set_index('DateTime', inplace=True)

# Resample to get daily average price and power consumption
daily_price = df['Price'].resample('D').mean()
daily_power = df['Power_Consumption'].resample('D').sum()  # / 1000 : Convert to MWh if original data is in kWh

# Find the peak and trough for price and power consumption
highest_price = daily_price.idxmax()
lowest_price = daily_price.idxmin()
highest_power = daily_power.idxmax()
lowest_power = daily_power.idxmin()

# Create the plot
fig, ax1 = plt.subplots(figsize=(15,7))

# Plot daily average price
color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (Euro/MWh)', color=color)
ax1.plot(daily_price.index, daily_price, label='Daily Average Price', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.scatter(highest_price, daily_price[highest_price], color='red', zorder=5)
ax1.scatter(lowest_price, daily_price[lowest_price], color='green', zorder=5)
ax1.annotate(f'{highest_price.strftime("%Y-%m-%d")}', (mdates.date2num(highest_price), daily_price[highest_price]), textcoords="offset points", xytext=(0,10), ha='center')
ax1.annotate(f'{lowest_price.strftime("%Y-%m-%d")}', (mdates.date2num(lowest_price), daily_price[lowest_price]), textcoords="offset points", xytext=(0,-15), ha='center')

# Create a secondary y-axis for daily power consumption
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Power Consumption (MWh)', color=color)
ax2.plot(daily_power.index, daily_power, label='Daily Power Consumption', color=color, linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)
ax2.scatter(highest_power, daily_power[highest_power], color='red', zorder=5)
ax2.scatter(lowest_power, daily_power[lowest_power], color='green', zorder=5)
ax2.annotate(f'{highest_power.strftime("%Y-%m-%d")}', (mdates.date2num(highest_power), daily_power[highest_power]), textcoords="offset points", xytext=(0,10), ha='center', color='red')
ax2.annotate(f'{lowest_power.strftime("%Y-%m-%d")}', (mdates.date2num(lowest_power), daily_power[lowest_power]), textcoords="offset points", xytext=(0,-15), ha='center', color='green')

# Legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Improve formatting
fig.autofmt_xdate()  # Auto rotates date labels
plt.title('Daily Average Price and Power Consumption with Highest and Lowest Points')
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlap

# Save the plot
plt.savefig('figures/powerprice2020.png')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data
df = pd.read_csv('data/yearlyData/commercial_df_2020.csv')

# Ensure that DateTime is a datetime type
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Set DateTime as the index
df.set_index('DateTime', inplace=True)

# Resample to get daily values
daily_power = df['Power_Consumption'].resample('D').sum() # / 1000  # Convert to MWh if original data is in kWh
daily_temp = df['Dew Point Temperature_P39'].resample('D').mean()
daily_humidity = df['Relative Humidity_P6'].resample('D').mean()
daily_wind_speed = df['Wind Speed_P4'].resample('D').mean()

# Create subplots
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 10), sharex=True)

# Plot daily power consumption
ax1.plot(daily_power.index, daily_power, label='Daily Power Consumption (MWh)', color='tab:blue')
ax1.set_ylabel('Power (MWh)')
ax1.legend(loc='upper left')
ax1.grid(True)

# Plot daily temperature
ax2.plot(daily_temp.index, daily_temp, label='Daily Average Temperature (C)', color='tab:red')
ax2.set_ylabel('Temperature (C)')
ax2.legend(loc='upper left')
ax2.grid(True)

# Plot daily humidity
ax3.plot(daily_humidity.index, daily_humidity, label='Daily Average Humidity (%)', color='tab:green')
ax3.set_ylabel('Humidity (%)')
ax3.legend(loc='upper left')
ax3.grid(True)

# Plot daily wind speed
ax4.plot(daily_wind_speed.index, daily_wind_speed, label='Daily Average Wind Speed (m/s)', color='tab:purple')
ax4.set_ylabel('Wind Speed (m/s)')
ax4.legend(loc='upper left')
ax4.grid(True)

# Formatting the date on the x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))  # Adjust the interval to display a date every 10 days
plt.gcf().autofmt_xdate()  # Auto-rotate the date labels

# Overall graph titles and layout adjustments
plt.suptitle('Daily Power Consumption and Weather Parameters')
plt.xlabel('Date')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust the rect to leave space for the suptitle

# Save the plot
plt.savefig('figures/daily_power_weather_2020.png')

# Show the plot
plt.show()
```
