###### When you're building a forecasting model, evaluating its performance over a single forecast horizon (such as one hour ahead) may not provide enough information about how well the model will perform in real-world scenarios. It's often more informative to evaluate the model over multiple future time points to see how it performs over time. For example, let's say you have hourly data, and youâ€™re interested in forecasting the power consumption for the next hour. Technically, you could set the test size to 1 when splitting your dataset, which means you'd be testing the model's prediction for just one hour ahead. However, this doesn't tell you much about the model's performance beyond that specific hour. Instead, you might choose to test the model's predictions for the next 24 hours. This way, you can see how well the model does not just one hour ahead, but two hours ahead, three hours ahead, and so on, up to 24 hours ahead. This gives you a much clearer picture of how the model performs over a range of future times, which is likely to be more representative of how the model will perform in practice.

```python
import os
# Assume kvarnholmen_df has columns 'DateTime', 'CUSTOMER', and 'Power_Consumption'
# Convert 'DateTime' to pandas datetime if it's not already
#kvarnholmen_df['DateTime'] = pd.to_datetime(kvarnholmen_df['DateTime'])
#kvarnholmen_df.set_index('DateTime', inplace=True)

# Select data for a single customer
customer_id = 1060753805  # Replace with actual customer ID
customer_data = kvarnholmen_df[kvarnholmen_df['CUSTOMER'] == customer_id]['Power_Consumption']  # y is your time series data with hourly readings, so here, y=customer_data which is the dataframe that includes power_consumption of customer 1060753805

# Ensure 'forecasts' directory exists
output_dir = "forecasts"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Define a function to perform the forecasting steps
def forecast_and_evaluate(data, frequency):
    # Resample data if frequency is not 'H' (hourly)  (your data is hourly by default)
    if frequency == 'H':  # Hourly
      sp = 24  # Assuming daily seasonality for hourly data (If the original data is hourly and you're expecting daily seasonality, you would set sp=24 for the hourly frequency.)
    elif frequency == 'D':  # Daily
      data = data.resample('D').sum()
      sp = 7  # Assuming weekly seasonality for daily data
    elif frequency == 'W':  # Weekly
      data = data.resample('W').sum()
      sp = None  # Seasonality might not be clear for weekly data
    elif frequency == 'M':  # Monthly
      data = data.resample('M').sum()
      sp = None  # Seasonality might not be clear for monthly data
    
    # You are interested in forecasting the next hour, but for evaluation, you want to see how the model performs over the next 24 hours
    test_size = 24  # 24 hours
    
    # Split the data
    y_train, y_test = temporal_train_test_split(data, test_size=test_size) # split data into train and test: total: 8760 (365*24) train: 8736 test:24
     
    # Define the forecasting horizon to be the next 24 hours
    fh = ForecastingHorizon(y_test.index, is_relative=False)   # fh=24 hours (it is for future, i.e., the test_size determines how many future time points you want to withhold from your training data to use for evaluation. The ForecastingHorizon (fh) is then used to specify which time points in the future you want to predict for.)
    
    # Train the model
    model = AutoARIMA(sp=sp, suppress_warnings=True)  # Seasonal periodicity (sp) can vary based on (input data) frequency
    
    # Now you can fit a forecasting model on y_train and evaluate its predictions using y_test, which contains the actual values for the next 24 hours.
    model.fit(y_train)
    
    # Forecast
    y_pred = model.predict(fh)
    
    # Evaluate
    mape = mean_absolute_percentage_error(y_pred, y_test)
    print(f'MAPE for frequency {frequency}: {mape}')
    
    # Plot
    plot_series(y_train, y_test, y_pred, labels=["y_train", "y_test", "y_pred"])
    plt.title(f"Forecasting at {frequency} frequency for customer {customer_id}")
    
    # Save the plot to the 'forecasts' directory
    plt.savefig(f'{output_dir}/customer_{customer_id}_forecast_{frequency}.png')
    
    # Show the plot
    plt.show()

# Forecast and evaluate for different frequencies (this means you make the input data hourly (if it is recorded in minute), daily, weekly, monthly)
for frequency in ['H', 'D', 'W', 'M']:
    print(f"Forecasting at {frequency} frequency for customer {customer_id}:")
    forecast_and_evaluate(customer_data.copy(), frequency)
```

```python
# fit an ARIMA model and plot residual errors
from datetime import datetime
from pandas import read_csv
from pandas import DataFrame
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot

# Define the date_parser function
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

# Read the CSV file while parsing the 'DateTime' column as dates
series = read_csv('data/commercial_df_2021.csv', header=0, parse_dates=['DateTime'], index_col='DateTime', date_parser=parser)

# Select the 'Power_Consumption' column and make sure it is in float format
series = series['Power_Consumption'].astype(float)

series.index = series.index.to_period('M')

# fit model
model = ARIMA(series, order=(5,1,0))  # In the ARIMA(5,1,0) model, p is 5, which means that the model uses the five previous time points (lags) to predict the next time point.
model_fit = model.fit()

# summary of fit model
print(model_fit.summary())

# line plot of residuals
residuals = DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()

# density plot of residuals
residuals.plot(kind='kde')
pyplot.show()

# summary stats of residuals
print(residuals.describe())
```
